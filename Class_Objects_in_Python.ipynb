```python
class Rectangle:
  def __init__ (self, length, width):
    self.length = length
    self.width = width
  def perimeter(self):
    x = 2 * (self.width + self.length)
    return x
  def area(self):
    y = self.width * self.length
    return y
my_rectangle = Rectangle(3, 4)
print(my_rectangle.area())
print(my_rectangle.perimeter())
```

```python
class Circle:
  def __init__ (self, radius, center):
    self.radius = radius
    self.center = center
  def area(self):
    x = 3.14 * self.radius
    return x
  def perimeter(self):
    y = 2 * 3.14 * self.radius
    return y
  def Inside(self):
    x = int(input())
    y = int(input())
    d = self.radius**2 - ((self.center-x)**2 + (self.center-y)**2)
    if d > self.radius:
      print('The point is not inside the circle.')
    elif d < self.radius:
      print('The Point is inside the circle. ')
    else:
      print('The Point touches the circle.')
my_circle = Circle(5, 10)
print(my_circle.area())
print(my_circle.perimeter())
my_circle.Inside()
```
